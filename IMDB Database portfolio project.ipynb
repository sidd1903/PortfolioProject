```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

df = pd.read_csv(r'C:\Users\sriva\Downloads\movies.csv')
```

```python
df.head()
```

```python
# looping through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#droping the rows with null values
df = df.dropna()
```

```python
# Checking the Data Types for our columns

print(df.dtypes)
```

```python
df.drop_duplicates()
```

```python
# sorting our Data for a better understanding

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#Plotting Regression Line against Gross and Budget to check correlation
sns.regplot(x="gross", y="budget", data=df)
```

```python
#Plotting Regression Line against Gross and Score to check correlation
sns.regplot(x="score", y="gross", data=df)
```

```python
# Correlation Matrix between all numeric columns

df.corr(method ='pearson',numeric_only=True)
```

```python
df.corr(method ='kendall',numeric_only=True)
```

```python
df.corr(method ='spearman',numeric_only=True)
```

```python
#Creating Heat map for Visual Representation of Correlation matrix
correlation_matrix = df.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_matrix = df.corr(numeric_only=True)

corr_pairs = correlation_matrix.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# Looking at the top 15 compaies by gross revenue

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
df.groupby(['company', 'year'])[["gross"]].sum()
```

```python
# Looking at the top 15 compaies by gross revenue by year

CompanyGrossSum = df.groupby(['company', 'year'])[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values(['gross','company','year'], ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```
